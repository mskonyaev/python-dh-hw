```python
# Задание 1
# В цикле использованы try/except для того, чтобы остановиться на последней биграмме, иначе будет out of range
text = '''варкалось хливкие шорьки пырялись по наве и хрюкотали зелюки как мюмзики в мове о бойся 
бармаглота сын он так свирлеп и дик а в глyще рымит исполин злопастный брандашмыг'''
text = text.split()
spisok_bigramm = []
i = 1
for word in text:
    try:
        word2 = text[i]
        i += 1
        spisok_bigramm.append((word, word2))     
    except: break
print(spisok_bigramm)   
```

```python
# Задание 2
import os

os.chdir(r'C:\Users\Максим\Documents\python-dh-hw\poems')
# Распечатать текущую директорию и список всех файлов и папок в ней
# Посчитать и распечатать количество файлов
c = 0
os.getcwd()
for file in os.listdir('.'):
    print(file)
    if os.path.isfile(file) == True:
        c += 1
print('Количество файлов: %d' % c) #Указано только количество файлов, без папок (в соответсвии с текстом задания!)
# Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt
# По завершении цикла должен получиться файл со всеми стихотворениями! (all_poems.txt)
for root, dirs, files in os.walk(r'C:\Users\Максим\Documents\python-dh-hw\poems'):
    for name in files:
        fullname = os.path.join(root, name)
        with open(fullname, 'r', encoding='utf-8') as f:
            text = f.read()
        with open('all_poems.txt', 'a', encoding='utf-8') as new_file:
            new_file.write(text)
            new_file.write('\n')
    
```

```python
# Задание 3
from string import punctuation

text = '''Удивительная кошка

Несчастная кошка порезала лапу-
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!'''

# 2. Узнать длину текста в символах
print(len(text))

# 3. Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов
new_text = [word.strip(punctuation) for word in text.split()]
print(new_text)

# 4. Объединить слова из получившегося списка в строку через пробел
line = ''
for word in new_text:
    line = (line + word + ' ')
print(line)
    
# 5 С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его
# к нижнему регистру. На выходе должен получиться список из всех слов стихотворения, написанных с маленькой
# буквы (не только те, которые изначально были с маленькой буквы, а все).
new_text = [word.lower() for word in new_text]
print(new_text)

# 6 Узнать длину текста в словах
print(len(new_text))

# 7 Узнать количество уникальных слов
print(len(set(new_text)))

# 8 Распечатать последние 10 слов
print(new_text[len(new_text)-10:len(new_text)])

# 9 Посчитать биграммы
## Способ N1

print(len(new_text) - 1)

## Способ N2
spisok_bigramm = []
i = 1
for word in new_text:
    try:
        word2 = new_text[i]
        i += 1
        spisok_bigramm.append((word, word2))     
    except: break
print(len(spisok_bigramm))

# 10 Соединить полученный список в строку. Слова в строке должны быть написаны через знак переноса строки
new_line = ''
for word in new_text:
    new_line = (new_line + word + '\n')
print(new_line)

# 11 Распечатать с 20 по 30 символ этой строки (включительно)
print(new_line[19:29])    
```
